```python
#  Import and read the charity_data.csv.
batter_df_2023 = pd.read_csv("Resources/fg_leaderboard_23.csv")
batter_df_2023 = batter_df_2023.rename(columns={'playerId': 'IDfg'})
batter_df_2023.head()
```

```python
for_2023_merge = batter_df_2023[['wOBA', 'SLG', 'BABIP', 'wRC+', 'IDfg']]

rename_columns = {
    'wOBA': 'wOBA_2023',
    'SLG': 'SLG_2023',
    'wRC+': 'wRC+_2023',
    'BABIP': 'BABIP_2023'
}

for_2023_merge.rename(columns=rename_columns, inplace=True)

for_2023_merge.head()
```

```python
players_batting_output = pd.merge(
    new_batter_df,
    for_2023_merge,
    on=['IDfg']
)
```

```python
avg_woba = players_batting_output['avg_wOBA']
avg_slg = players_batting_output['SLG']
avg_babip = players_batting_output['BABIP']
avg_wrc = players_batting_output['wRC+']

woba_2023 = players_batting_output['wOBA_2023']
slg_2023 = players_batting_output['SLG_2023']
babip_2023 = players_batting_output['BABIP_2023']
wrc_2023 = players_batting_output['wRC+_2023']

z_scores_avg_woba = zscore(avg_woba)
z_score_woba_2023 = zscore(woba_2023)

z_scores_avg_slg = zscore(avg_slg)
z_score_slg_2023 = zscore(slg_2023)

z_scores_avg_babip = zscore(avg_babip)
z_score_babip_2023 = zscore(babip_2023)

z_scores_avg_wrc = zscore(avg_wrc)
z_score_wrc_2023 = zscore(wrc_2023)

# Add new columns with Z-scores to the DataFrame
players_batting_output['z_scores_avg_woba'] = z_scores_avg_woba
players_batting_output['z_scores_avg_slg'] = z_scores_avg_slg
players_batting_output['z_scores_avg_babip'] = z_scores_avg_babip
players_batting_output['z_scores_avg_wrc+'] = z_scores_avg_wrc

players_batting_output['zscore_difference_woba'] = z_score_woba_2023 - z_scores_avg_woba
players_batting_output['zscore_difference_slg'] = z_score_slg_2023  - z_scores_avg_slg
players_batting_output['zscore_difference_babip'] = z_score_babip_2023 - z_scores_avg_babip
players_batting_output['zscore_difference_wrc+'] = z_score_wrc_2023 - z_scores_avg_wrc
```

```python
def process_batting_data(input_df, target_column1, target_column2):
    zscore_columns = [target_column1, target_column2]
    
    for col in zscore_columns:
        z_scores = zscore(input_df[col])
        input_df[f'z_scores_{col}'] = z_scores

    zscore_difference_column = f'zscore_difference_{target_column2}'
    input_df[zscore_difference_column] = input_df[f'z_scores_{target_column2}'] - input_df[f'z_scores_{target_column1}']
    
    return input_df
```
